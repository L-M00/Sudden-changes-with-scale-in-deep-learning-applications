```python
import cellpylib as cpl
```

```python
#@title Cellular automata test
rule_number = 90
cellular_automaton = cpl.init_simple(200) #creates a list of a list of all zeros of length 200
cellular_automaton = cpl.evolve(cellular_automaton, timesteps=100, memoize=True, apply_rule=lambda n, c, t: cpl.nks_rule(n, rule_number))

#print(cellular_automaton[-1]) #using [-1] gives the final element of a list
cpl.plot(cellular_automaton, title='Rule {0}, periodic boundary conditions'.format((rule_number)))
```
