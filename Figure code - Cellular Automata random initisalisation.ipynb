```python
rule_number = 30
cellular_automaton = cpl.init_random(100) #creates a list of a list of all zeros of length 200
cellular_automaton = cpl.evolve(cellular_automaton, timesteps=100, memoize=True, 
                                apply_rule=lambda n, c, t: cpl.nks_rule(n, rule_number))
fig1 = cpl.plot(cellular_automaton, title='Rule {0}'.format((rule_number)))
```
