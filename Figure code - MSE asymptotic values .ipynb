```python
n_list = np.arange(2, 22, 2)
# integral_list corresponds to the values of integrals between the nth Legendre polynomial and cos(pi x) between -1 and 1
integral_list = [-0.607927, 0.129433, -0.0101715, 0.000423335, -0.0000109711, 
                       1.94534*10**(-7), -2.51162*10**(-9), 2.47373*10**(-11), -7.64626*(10**-13), 4.10363*10**(-13)]
```

```python
beta = np.empty(len(n_list))
for i in range(len(beta)):
    beta[i] = integral_list[i] * (2*n_list[i] + 1)/2
print(beta)
```
